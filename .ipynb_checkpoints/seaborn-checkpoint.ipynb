```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.heatmap(df.corr(),annot=True)
```

# Joint Plot

### Bivariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pairplot

### Multivariate Analysis

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Distplot

### Univariate Analysis

```python
sns.distplot(df.tip)
```

```python
sns.distplot(df.tip,kde=False)
```

# Categorical Plots

# Count Plot

```python
sns.countplot(df.sex)
```

```python
sns.countplot(df.smoker)
```

```python
sns.countplot(df.day)
```

```python
sns.countplot(y=df.sex)
```

```python
sns.barplot(y=df.total_bill,x=df.sex)
```

## Boxplot

```python
sns.boxplot(df.smoker,df.total_bill,df.sex)
```

```python
sns.boxplot(df.smoker,df.total_bill,df.sex,palette='Set2')
```

```python
sns.boxplot(df.total_bill,df.sex)
```

```python
sns.boxplot(data=df)
```
