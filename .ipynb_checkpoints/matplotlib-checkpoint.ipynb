```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
y=x*x
```

```python
plt.plot(x,y,'ko--')
plt.show()
```

```python
plt.plot(x,y,linestyle=':',linewidth=10,markersize=100)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Squared Power")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,c='k')
plt.subplot(2,2,3)
plt.plot(x,y,'g--')
plt.subplot(2,2,4)
plt.plot(x,y,":")
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.title("Sine wave")
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,1,1)
x=np.arange(0,10,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'r:')
plt.plot(x,z,'c--')
```

```python
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,16,7]
plt.bar(x2,y2,color='c')
plt.bar(x,y,color='r')
```
